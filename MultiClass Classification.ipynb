```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# make 4-class dataset for classification
classes = 4
m = 100
centers = [[-6, 2], [-2, -2], [1, 2], [5, -2]]
std = 1.0
X_train, y_train = make_blobs(n_samples=m, centers=centers, cluster_std=std,random_state=30)
np.unique(y_train)
```

```python
fig, ax = plt.subplots()
y = [zero, one, two, three] = [y_train == 0, y_train == 1, y_train == 2, y_train == 3]
ax.scatter(X_train[zero][:,0],X_train[three][:,1], c='g', marker = 's', label='y=0')
ax.scatter(X_train[one][:,0],X_train[one][:,1], c='b', marker = 'o', label='y=1')
ax.scatter(X_train[two][:,0],X_train[two][:,1], c='r', marker = 'x', label='y=2')
ax.scatter(X_train[three][:,0],X_train[three][:,1], c='y', marker = '^', label='y=3')
plt.legend()
plt.show()
```

```python
model = Sequential([
    Dense(units=2, activation='relu'),
    Dense(units=4, activation='linear')
])
```

```python
model.compile(
    loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer= tf.keras.optimizers.Adam(0.01),
)
model.fit(X_train, y_train, epochs=200)
```
